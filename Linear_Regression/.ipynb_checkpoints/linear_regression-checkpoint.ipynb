```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('homeprices.csv')
df
df.columns
    
      
```

In Jupyter Notebooks, %matplotlib inline is a magic command that allows the output of plotting commands to be displayed directly within the notebook. This line is specific to Jupyter Notebooks and is not a standard Python code.

When %matplotlib inline is used, it sets the backend of matplotlib to the 'inline' backend, which means that the plots will be rendered in the output cells of the Jupyter Notebook. This is particularly useful when you're working on data visualization tasks and want to see the plots directly in the notebook rather than in separate windows.

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')

plt.scatter(df.area,df.price,color='red',marker='+')
```

```python


reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
input_scalar = 3300.0
input_array = np.array(input_scalar).reshape(1, -1)

reg.predict(input_array)
```
